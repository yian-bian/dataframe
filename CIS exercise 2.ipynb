```python
# get the original data
import pandas as pd
import yfinance as yf
table = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
df = table[0]
# get a list of tickers
stockdata = df['Symbol'].to_list()
# get the data of the stocks of tickers that we want from 2022-01-21 to 2022-02-12
full_stock_data = yf.download(stockdata, '2022-01-21', '2022-02-12')
full_stock_data
```

```python
# A. DATA. Choose a universe with at least 10 names. Remember our START DATE 
# from Exercise 1. Take the price of every stock in the universe on the START DATE as well as ONE WEEK prior and every multiple of ONE WEEK thereafter. 

# get the data we want --- the adjusted close price for each stock
stock_adjclose = full_stock_data['Adj Close']
# get the stock price at end of every week(as every weekend the stock market closed, we actually get the stock price of every Friday)
price_at_the_end_of_the_week = stock_adjclose.resample('W').ffill()
price_at_the_end_of_the_week
```

```python
# B. SIGNAL GENERATION. Compute all the one week returns for each name. These are the signals for our momentum strategy. 
# Provide a table of the signals (ticker, week)

# calculate the weekly return of the stock adjusted close
weekly_return = stock_adjclose.resample('W').ffill().pct_change() 
# as the first row would be NA, we delete the first row
weekly_return = weekly_return[1:]         
# generate a Series that we want to use as index
date = pd.Series(['week1','week2','week3'])    
# change the index for the dataframe to what we want
weekly_return = weekly_return.set_index(date)                       
weekly_return
```

```python
# C. PORTFOLIO CONSTRUCTION. Assume we go long the top quintile and short the bottom quintile.
# Given a Target Notional of $2 Million of gross exposure (long and short), use signal weighting to compute the positions each week. 
# Bonus: Note that this simple portfolio construction is not perfectly hedged. What might you suggest to improve that? 

# in order to hedge the portfolio, maybe we can be more diversify, as stocks in one industry may all do well or bad,
# so maybe we can apply signal weighting separately in different industry and combine 
# the stocks that do top and last in its industry to contruct out portfolio

# signal weighting using week 1 data

# transpose the dataframe
df = weekly_return.T    
# sort the data by the order of week1 reuturn
week_1 = df.sort_values(by = 'week1', ascending = False)      
# get the top100 stocks to long
week_1_long = week_1.head(100) 
# get the sum of the return of the stocks that we long to be the denominator of the signal weighting
week_1_long = week_1_long.assign(sum1 = week_1_long.week1.sum())    
# use every stock's return to divided by the sum of all the returns to get their position 
week_1_long['signal_weighting_long'] = week_1_long.week1/week_1_long.sum1                   
# use their return to multiple their position to see if invest $1, how much money will each stock get as the return
week_1_long = week_1_long.assign(week2return = week_1_long.signal_weighting_long * week_1_long.week2)

# get the last 100 stocks to short
week_1_short = week_1.tail(100)   
# get the sum of the return of the stocks that we long to be the denominator of the signal weighting
week_1_short = week_1_short.assign(sum2 = week_1_short.week1.sum()) 
# use every stock's return to divided by the sum of all the returns and take the negative to get their position,as we short the data 
# so we should use negative position
week_1_short['signal_weighting_short'] = -week_1_short.week1/week_1_short.sum2 
# # use their return to multiple their position to see if invest $1, how much money will each stock get as the return
week_1_short =week_1_short.assign(week2return = week_1_short.signal_weighting_short * week_1_short.week2)



# signal weighting using week 2 data, the same process as how we do it for week 1
week_2 = df.sort_values(by = 'week2', ascending = False)
week_2_long = week_2.head(100)
week_2_long = week_2_long.assign(sum1 = week_2_long.week2.sum())
week_2_long['signal_weighting_long'] = week_2_long.week1/week_2_long.sum1
week_2_long = week_1_long.assign(week3return = week_2_long.signal_weighting_long * week_2_long.week3)

week_2_short = week_2.tail(100)
week_2_short = week_2_short.assign(sum2 = week_2_short.week2.sum())
week_2_short['signal_weighting_short'] = -week_2_short.week2/week_2_short.sum2
week_2_short =week_2_short.assign(week3return = week_2_short.signal_weighting_short * week_2_short.week3)
```

```python
# D. TRADE EXECUTION. What transaction costs should we assume? 
# fee cost: 0.2bps, usually 0.1~0.2bps over 1-2M
# latency cost: as this is a relatively short-term trade, there definitely will have latency cost, just assume it to be 0.1bp
# fill cost: as we have 200 stocks here, the average fill cost is 0.5bps(include opportunity cost).0.15-1.5bps, as we use as an average, so we use 0.5bps
```

```python
# E. BACKTEST. Finally, calculate the weekly returns of this strategy. Sum these to get the total return of the strategy. 

transaction_cost_pct = (0.2+0.1+0.5) * (0.0001)
# after using week 1's data to calculate the signal weightig, and using that sizing strategy, we have return of week 2:
week2Return = 1000000*(week_1_long['week2return'].sum() + week_1_short['week2return'].sum())
# after week 2, we have :
after_week2 = 2000000 + week2Return
# the money we have to enter week 3:
enter_week3 = after_week2 * (1 - transaction_cost_pct)
# after week 3, we have:
week3Return = (enter_week3/2)*(week_2_long['week3return'].sum() + week_2_short['week3return'].sum())
after_week3 = week3Return + enter_week3
# the money we have intotal to enter week 4:
enter_week4 = after_week3 * (1- transaction_cost_pct)
enter_week4
```
